# 五、如何确定模型的阶数？

### 核心问题是，我们如何确定两个量是相关的？  
我们用方差衡量波动：  
$Var(x) = \frac{1}{n} \sum_{i}^{} (x_i-\mu)^2$
用均值衡量期望：  
$E(x) = \frac{1}{n} \sum_i x_i$
### 一种非常原始的思路是，分情况讨论：  
对于随机变量$X,Y$的一次抽样$x_i,y_i$  
如果$x_1$和$x_2$乃至$x_i$对应的 $E(Y)$ 都近似相等甚至完全相等，我们就有充分的证据认为$X,Y$独立  
为了让$E(Y)$的对比更简单，我们将$Y$做平移变换：  
# $Y'=Y-E(Y)$  
这样有  
# $E(Y') = 0$
如果$X,Y$独立的话，会有  
# $E(x_iE(Y')) = 0$
也就是  
# $\sum_i x_iE(Y') = 0$  
为了减小计算量(由$n^2$次到$n$次)，我们计算  
# $\sum_i x_i(y_i-\overline{y})$  
和$Y$保持一致，对$X$也做平移变换，可以得到  
# $\sum_i (x_i - \overline{x})(y_i-\overline{y})$
这已经非常接近最后的结果了。最后一步是取均值  
# $\frac{1}{n}\sum_i (x_i - \overline{x})(y_i-\overline{y})$  
我们称这个量为协方差，记为  
# $Cov(X,Y)=\frac{1}{n}\sum_i (x_i - \overline{x})(y_i-\overline{y})$
### 显然，当$X,Y$独立时，有$Cov$为$0$。简单推导可以知道，当$y=kx+b$时，如果$k$为正，有$Cov$为正，$k$为负，有$Cov$为负

# 但还有一个问题是，$X,Y$的波动可以不一致。弱相关强波动的$Cov$可能大于强相关弱波动的$Cov$，因此Cov不具有相互比较的意义，只有与0比较的意义。
# 解决的策略是，用标准差进行归一化

## $r = \frac{Cov(X,Y)}{\sqrt{Var(X)}\sqrt{Var(Y)}}$  
## 归一化之后，当$y=kx+b$时，如果$k$为正，有$r$为1，$k$为负，有$r$为-1

```python
def DistributionPlot(data,a,b,n):
    step = (b-a)/n
    Distrubiution = {}
    for j in data: #将Distrubiution数组初始化，即创建索引
        Distrubiution[j//step*step+0.5*step] = 0

    for j in data: #将数值转化为频次
        Distrubiution[j//step*step+0.5*step] += 1  
    
    Distrubiution = pd.Series(Distrubiution) # 用Series盛装，方便绘图
    plt.plot(Distrubiution)
    plt.show()
    return Distrubiution
```
